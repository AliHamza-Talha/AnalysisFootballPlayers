```python
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
data = pd.read_csv("dataset.csv")
```

```python
data.describe()
```

```python
data.head()
```

```python
#JE VEUX REALISER UNE ACP SUR CE DATASET POUR VOIR SI JE PEUX INTREPRETER LES RESULTATS
#POUR CELA JE RECUP QUE LES DONNEES NUMERIQUES

columns_to_drop = ['UID', 'Name', 'Born']
data_numeric = data.drop(columns=columns_to_drop)
data_numeric = data_numeric.select_dtypes(include=[float, int])

# Standardiser les données
scaler = StandardScaler()
df_scaled = scaler.fit_transform(data_numeric)

# Effectuer l'ACP avec deux composantes principales
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)

# Variance expliquée par chaque composante principale
explained_variance = pca.explained_variance_ratio_
print(f"Variance expliquée par la première composante: {explained_variance[0]:.2f}")
print(f"Variance expliquée par la deuxième composante: {explained_variance[1]:.2f}")

# Ajouter les résultats de l'ACP au DataFrame
df_pca = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])

# Visualiser les deux premières composantes principales
plt.figure(figsize=(8,6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.5)
plt.title('ACP: Projection des données sur les deux premières composantes principales')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.show()
```

```python
# Calculer la variance expliquée cumulée
pca = PCA()
pca_result = pca.fit_transform(df_scaled)
explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Visualiser la variance expliquée cumulée
plt.figure(figsize=(8,6))
plt.plot(np.arange(1, len(explained_variance)+1), explained_variance, marker='o', linestyle='--')
plt.xlabel('Nombre de composantes principales')
plt.ylabel('Variance expliquée cumulée')
plt.title('ACP : Variance expliquée cumulée en fonction du nombre de composantes')
plt.axhline(y=0.90, color='r', linestyle='--')  # Ligne rouge à 90% de variance expliquée
plt.grid(True)
plt.show()
```

```python
#--->ECHEC TOTAL DE L'ACP IL FAUDRAIT AVOIR UNE REPRESENTATION EN DIMENSION 40 ENVIRON POUR COMMENCER A INTERPRETER LA VISUALISATION
#ON VA QUAND MM REGARDER CE QUI SE PASSE EN 3 DIMENSION

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardiser les données
scaler = StandardScaler()
df_scaled = scaler.fit_transform(data_numeric)

# Effectuer l'ACP avec 3 composantes principales
pca = PCA(n_components=3)
pca_result = pca.fit_transform(df_scaled)

# Créer un DataFrame pour les trois premières composantes
df_pca_3d = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2', 'PC3'])

# Visualiser en 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Tracer les points
ax.scatter(df_pca_3d['PC1'], df_pca_3d['PC2'], df_pca_3d['PC3'], alpha=0.5)

# Ajouter des étiquettes
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('ACP : Projection des données sur les trois premières composantes')

plt.show()
```
